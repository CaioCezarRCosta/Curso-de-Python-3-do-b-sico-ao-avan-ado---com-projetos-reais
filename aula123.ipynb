Métodos úteis dos dicionários em Python 

- len - Quantas chaves 
- keys - Iterável com as chaves 
- values - Iterável com os valores 
- items - Iterável com as chaves e valores 
- setdefault - Adiciona valor se a chave não existe
- copy - Retorna uma cópia rasa (Shallow copy)
- get - Obtém a chave 
- pop - Apaga um tem com a chave especificada (del)
- popitem - Apaga o último item adicionado 
- update - Atualiza um dicionário com outro 

Dicionário exemplo


```python
d1 = {
    'c1': 1,
    'c2': 2
}
d2 = d1

d2['c1'] = 1000
print(d1)

#Não é pra copiar dessa forma
```

copy

```python
d1 = {
    'c1': 1,
    'c2': 2,
    'l1': [0,1,2]
}
d2 = d1.copy()

d2['c1'] = 1000
d2['l1'][1] = 999999999
print(f'Dicionário d1: {d1}')
print(f'Dicionário d2 {d2}')

#Não é pra copiar dessa forma
#Copiam valores imutáveis e linka os mutáveis
```

deepcopy

```python
import copy
d1 = {
    'c1': 1,
    'c2': 2,
    'l1': [0,1,2],
}

d2 = copy.deepcopy(d1)

d2['c1'] = 1000
d2['l1'][1] = 999999999
print(f'Dicionário d1: {d1}')
print(f'Dicionário d2 {d2}')

#Não é pra copiar dessa forma
#Copiam valores imutáveis e mutáveis
```
