# 🐍 Dicionários em Python (`dict`)

Os **dicionários** são estruturas do tipo **“chave” e “valor”**.

- As **chaves** funcionam como os **índices** das listas, mas precisam ser de **tipos imutáveis**, como:
  - `str`, `int`, `float`, `bool`, `tuple`, etc.
- Os **valores** podem ser de **qualquer tipo**, incluindo **outros dicionários**.

Podemos criar dicionários usando **chaves `{}`** ou a **classe `dict()`**.

---

## 🔒 Tipos Imutáveis:
- `str`
- `int`
- `float`
- `bool`
- `tuple`

## 🔓 Tipos Mutáveis:
- `dict`
- `list`

---

## 💡 Exemplo de Dicionário:

```python
pessoa = {
    'nome': 'Luiz Otávio',
    'sobrenome': 'Miranda',
    'idade': 18,
    'altura': 1.8,
    'enderecos': [
        {'rua': 'tal tal', 'número': 123},
        {'rua': 'outra rua', 'número': 321},
    ]
}


Criação do dicionário Pessoa

````python
pessoa = {
    'nome': 'Luiz Otávio',
    'sobrenome': 'Miranda',
    'idade': 18,
    'altura': 1.8,
    'enderecos': [
        {'rua': 'tal tal', 'número': 123},
        {'rua': 'outra rua', 'número': 321},
    ]
}
print(pessoa, type(pessoa))
````

Assessando itens de um dicionário

````python
print(pessoa['nome'])
print(pessoa['sobrenome'])
````

Percorrendo um dicionário com o for

````python
for chave in pessoa:
    print(chave, pessoa[chave])
````

Formas de criar um Dicionário

````python
pessoa_aula = {
    'nome': 'Caio Cezar',
    'sobrenome' : 'Ribeiro Costa'
}

pessoa_aula_dict = (dict(nome='Caio Cezar ', sobrenome= 'Ribeiro Costa'))
print(type(pessoa_aula_dict))
print(type(pessoa_aula))
````
