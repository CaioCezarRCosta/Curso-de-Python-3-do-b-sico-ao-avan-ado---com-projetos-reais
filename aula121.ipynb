# 🐍 Dicionários em Python (`dict`)

Os **dicionários** são estruturas do tipo **“chave” e “valor”**.

- As **chaves** funcionam como os **índices** das listas, mas precisam ser de **tipos imutáveis**, como:
  - `str`, `int`, `float`, `bool`, `tuple`, etc.
- Os **valores** podem ser de **qualquer tipo**, incluindo **outros dicionários**.

Podemos criar dicionários usando **chaves `{}`** ou a **classe `dict()`**.

---

## 🔒 Tipos Imutáveis:
- `str`
- `int`
- `float`
- `bool`
- `tuple`

## 🔓 Tipos Mutáveis:
- `dict`
- `list`

---

## 💡 Exemplo de Dicionário:

```python
pessoa = {
    'nome': 'Luiz Otávio',
    'sobrenome': 'Miranda',
    'idade': 18,
    'altura': 1.8,
    'enderecos': [
        {'rua': 'tal tal', 'número': 123},
        {'rua': 'outra rua', 'número': 321},
    ]
}


````python
pessoa = {}

pessoa['nome'] = 'Caio Cezar'
print(pessoa)
print(pessoa['nome'])
````

Erro: Chave inexistente

````python
print(pessoa['nome1'])
````

Erro index inexistente

````python
lista=[]
print(lista['nome1'])
````

Criando chave dinámica para um dicionário

````python
pessoa = {}
chave = 'nome'
pessoa[chave]='Caio Cezar'
pessoa['sobrenome'] = 'Ribeiro'
print(pessoa[chave])
pessoa[chave] = 'Maria'
print(pessoa)
````

Deletando chave de um dicionário

````python
del pessoa['sobrenome']
print(pessoa)
print(pessoa['sobrenome'])
````

Formas de verificar que existe (não dunciona)

````python
if pessoa['sobrenome']:
    print('Existe')
else:
    print('Não existe')
````

````python
print(f'Passando um "else" : {pessoa.get('sobrenome','Não existe')}')
#Se passar como parâmetro somente a chave, e não o "else", ele retorna por padrão o none
print(f'Sem passar o "else": {pessoa.get('sobrenome')}')
````
